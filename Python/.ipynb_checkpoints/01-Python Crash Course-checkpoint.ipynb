```python
operator = input("Choose one of the operator (+,-,*,/,%,**): ")
number1= int(input("Enter the first number: "))
number2= int(input("Enter the second number: "))

if operator == '+':
    print(number1+number2)
elif operator == "-":
    print(number1-number2)
elif operator == '*':
    print(number1*number2)
elif operator == '/':
    print(number1/number2)
elif operator == '%':
    print(number1%number2)
elif operator == '**':
    print(number1**number2)
else:
    print("Invalid operator or number")
```
